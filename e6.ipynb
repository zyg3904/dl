```python
# Import the required libraries
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 2: Load the YOLO model
model = YOLO('yolov5su.pt')
```

```python
print("Please upload an image file to analyze.")
uploaded = files.upload()
image_path = next(iter(uploaded))
image = cv2.imread(image_path)

# Check if the image file could be read
if image is None:
    raise Exception(f"\nError: Could not load image from '{image_path}'. The file may be corrupted or unsupported.")


print(f"\nSuccessfully loaded '{image_path}'. Processing...")

# Convert the image color from BGR (OpenCV) to RGB (matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Perform object detection
results = model(image_rgb)

# Get the image with bounding boxes drawn on it
annotated_frame = results[0].plot()

# Display the final image
print("Displaying detection results...")
plt.figure(figsize=(12, 10))
plt.imshow(annotated_frame)
plt.title("YOLOv5 Object Detection Result")
plt.axis('off')
plt.show()
```
